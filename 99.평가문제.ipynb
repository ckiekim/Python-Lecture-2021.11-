### Q-14
ì •ìˆ˜ numì„ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ì•„ ê° ìë¦¬ ìˆ«ì(digit)ì˜ í•©ì„ ê³„ì‚°í•˜ëŠ” sumOfDigits() í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ê³ , ì˜ˆë¥¼ ë“¤ì–´ ê²€ì¦í•´ ë³´ì„¸ìš”.

```python
def sumOfDigits(num):
    s = str(num)
    sum_of_digits = 0
    for digit in list(s):
        sum_of_digits += int(digit)
    return sum_of_digits

print(sumOfDigits(123456))
```

1. ì°¾ê³ ì í•˜ëŠ” ë²”ìœ„ì˜ ìì—°ìˆ˜ë¥¼ ë‚˜ì—´í•œë‹¤.
2. 2ë¶€í„° ì‹œì‘í•˜ì—¬, 2ì˜ ë°°ìˆ˜ë¥¼ ì§€ì›Œë‚˜ê°„ë‹¤.
3. ë‹¤ìŒ ì†Œìˆ˜ì˜ ë°°ìˆ˜ë¥¼ ëª¨ë‘ ì§€ìš´ë‹¤.
</pre>
ì´ì™€ ê°™ì€ ë°©ì‹ìœ¼ë¡œ 100 ì´í•˜ì˜ ì†Œìˆ˜ë¥¼ êµ¬í•˜ì„¸ìš”.

```python
def is_prime(n):
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
```

```python
prime_numbers = [i for i in range(2, 201) if is_prime(i)]
print(prime_numbers)
```

```python
numbers = list(range(2,101))
for i in prime_numbers:
    for k in range(2*i,101,i):
        if k in numbers:
            numbers.remove(k)
print(numbers)
```

### Q-19
ìƒê·¼ì´ê°€ ë§ì¶° ë†“ì€ ì•ŒëŒ ì‹œê°ì„ ì…ë ¥ìœ¼ë¡œ ë°›ì•„(ì‹œì™€ ë¶„) ì°½ì˜ì´ì˜ ë°©ë²•ì„ ì‚¬ìš©í•œë‹¤ë©´, ì´ë¥¼ ì–¸ì œë¡œ ê³ ì³ì•¼ í•˜ëŠ”ì§€ êµ¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.

```python
def early_alarm(hour, minute):
    new_hour, new_min = hour, minute-45
    if new_min < 0:
        new_min = 60 + minute - 45
        new_hour = hour - 1
    return new_hour, new_min

h, m = map(int, input('ì‹œ ë¶„').split())
new_hour, new_min = early_alarm(h, m)
print(f'{h:02d}:{m:02d} ==> {new_hour:02d}:{new_min:02d}')
```

### Q-20
https://www.lipsum.com/ ì‚¬ì´íŠ¸ì— ì ‘ì†í•´ì„œ ë³¸ë¬¸ì˜ ë‚´ìš©ì„ lipsum.txt íŒŒì¼ì— ì €ì¥í•˜ê³ , ì´ íŒŒì¼ì„ ì½ì–´ ë‹¨ì–´ì˜ ê°œìˆ˜ë¥¼ ì„¸ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.<br>ë‹¨, ì½ì€ ë‹¨ì–´ëŠ” ì†Œë¬¸ìë¡œ ë³€í™˜í•˜ê³ , êµ¬ë‘£ì ì„ ì—†ì•¤ í›„ ê°œìˆ˜ë¥¼ ì„¸ê³ ,<br>ì¶œë ¥ì€ ì´ ë‹¨ì–´ìˆ˜ì™€ ê°€ì¥ ë§ì´ ë‚˜ì˜¨ ìˆœì„œëŒ€ë¡œ ë‹¨ì–´ 10ê°œì™€ ê·¸ ë‹¨ì–´ì˜ ë¹ˆë„ì…ë‹ˆë‹¤.


```python
with open('lipsum.txt') as file:
    lines = file.readlines()

text = []
for line in lines:
    if line == '\n':
        continue
    text += line.split()

words = []
for word in text:
    word = word.strip(',.?\'\"()').lower()
    words.append(word)
```

```python
word_dict = {}
for word in words:
    if word in word_dict.keys():
        word_dict[word] += 1
    else:
        word_dict[word] = 1
```
